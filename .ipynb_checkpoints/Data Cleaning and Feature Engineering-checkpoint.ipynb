# Data Cleaning
 
Let's FIrst Look at the Data itself to get a idea of where we need to do work and how much feature tweaking is reauired:

### There are 4 Datasets availbale for this problem namely:
1. train.csv - This Dataset contains all the initial ticket information (all tickets issued 2004-2011)
2. validation.csv - this Dataset contians 61000 never seen instances for final evaluation of the trained model (all tickets issued 2012-2016)
3. addresses.csv - This Dataset contains additional address information of each ticket recipient
4. lat-lon.csv - mapping from ticket id to addresses, and from addresses to lat/lon coordinates. 

```python
#Loading Data and Dependencies

%matplotlib inline

import pandas as pd
import numpy as np
import missingno as ms

train = pd.read_csv('DataSet/train.csv',encoding='ISO-8859-1',low_memory=False)
adrs = pd.read_csv('DataSet/addresses.csv',encoding='ISO-8859-1',low_memory=False)
ltln = pd.read_csv('DataSet/lat-lon.csv',encoding='ISO-8859-1',low_memory=False)
```

## Looking at Train.csv

```python
print(train.columns)
train.head(3)
```

```python
#Checking Null Values
ms.bar(train);
```

```python
#As, we cam see that 5 columns (violation_zip_code,non_us_str_code,payment_date,collection_status,grafitti_status) 
#are more or less completely filled with NaN or empty thus dropping these columns

train.drop(['violation_zip_code','non_us_str_code','payment_date','collection_status','grafitti_status'],axis=1,inplace=True)

#And since we will be focusing only on the violator that were found responsible, removing all instances where violator
#was not responsible i.e where compliance is NaN

train.dropna(inplace=True)
train.head(3)
```

```python
ms.bar(train);
```

```python
ms.bar(ltln);
```

```python
adrs['address'] = adrs['address'].str.extract(r'[0-9]* (.*),')
ltln['address'] = ltln['address'].str.extract(r'[0-9]* (.*),')
print('Address in adrs DS: {} \nAddress in ltln DS: {}'.format(len(adrs.address.unique()),len(ltln.address.unique())))
```

# Feature Engineering:

The Datasets contains several discriptive and string based features from which, new and Machine Learning ready feature needs to be extracted.  <br />
New Feature will be formed and evaluated based on their respective co-relation with the target  <br /> 
### The Feature Contains two types of feature Values namely Categorical and Continuous
1. For Categorical data Dummy encoding or One Hot-key encoding will suffice
2. And Continuous Data will be Normalized Using MinMax Scaling

## Extracting New Features from Discriptive and String Based Features

```python
train.columns
```

## 1. Using Feature: violator_name

```python
#A Function to map Violator_type
def get_violator_feature(df):
    df['len_violator'] = df['violator_name'].apply(lambda x : len(str(x)) if 'Inc' not in str(x) and 'L.L.C' not in str(x) 
                                                 and 'INC' not in str(x) and 'LLC' not in str(x) else 0)
    df['violator_type'] = df['len_violator'].apply(lambda x : 'Agency' if x==0 or x >20 else 'Person' )
    return df
```

```python
train = get_violator_feature(train)
train.iloc[:10,[3,-1]]
```

## 2. Using Feature: violation_discription

```python
#A Function to map length and number of violations
def get_discription_feature(df):
    #Lenght of violation    
    df['len_description'] = df['violation_description'].str.len()
    #Number of violation
    df['count_violation'] = df['violation_description'].apply(lambda x : len(x.split(',')))
    return df
```

```python
train = get_discription_feature(train)
train.iloc[:10,[15,-2,-1]]
```

## 3. Using Feature: disposition

```python
# A Function to map Reason of default and if fine was waived or not
def get_disposition_features(df):
    # Reason of Guilt
    df['responsible_by'] = df['disposition'].apply(lambda x : 0 if 'Not responsible' in x else -1 if 'PENDING' in x else x.split()[-1])
    #Correcting Data Errors
    df.loc[df['responsible_by'] == 'Deter','responsible_by'] = 'Determination'
    df.loc[df['responsible_by'] == 'Admis','responsible_by'] = 'Admission'
    # If Fine was Waived or not
    df['fine_waived'] = df['disposition'].apply(lambda x : 1 if 'Fine Waived' in x else 0)
    return df
```

```python
train = get_disposition_features(train)
train.iloc[:10,[16,-2,-1]]
```

## 4. Using Feature: ticket_issued_date

```python
#A Function to map Month_bin and Time of ticket issued
def get_date_time_feature(df):
    #time of ticket_issued
    df['ticket_time'] = df['ticket_issued_date'].apply(lambda x : int(''.join(x[11:].split(':')[0:2])))
    #Month_bin
    df['month_bin'] = df['ticket_issued_date'].apply(lambda x : 1 if 1 <= int(x[5:7]) <= 3 else 2 if 4 <= int(x[5:7]) <= 6 
    else 3 if 7 <= int(x[5:7]) <= 9 else 4)
    return df
```

```python
train = get_date_time_feature(train)
train.iloc[:10,[12,-2,-1]]
```

## 5.  Adding Latitude and Longitude from add_coor Dataset

```python
#merging the Dataset
train = pd.merge(train,add_coor,how='left')
```

```python
#cheking for NaN values
ms.bar(train);
```

## Normalization Method Used :

<font size="4">To have all the feature values on the same scale we need to Noramlize all feature values to a scale of [0,2] :</font><br /><br />

<font size="4">Normalizing the Feature Value Score To a Range of [0,2] by :-</font>
<font size="5">
## <img src="https://render.githubusercontent.com/render/math?math=Value STD = (\frac{Feature Value - Min(Feature Value)}{Max(Feature Value) - Min(Feature Value)})">   

## <img src="https://render.githubusercontent.com/render/math?math=Normalized Value = Value STD*(2-0)%2B0"></font>

```python
#Selecting Columns to Normalize

print(train.columns)
display(train.iloc[:2,15:30])
display(train.iloc[:2,30:])
```

```python
# To Normalize: ['fine_amount', 'admin_fee', 'state_fee', 'late_fee', 'discount_amount',
#'clean_up_cost', 'judgment_amount', 'payment_amount', 'balance_due','len_violator','len_discription']

from sklearn.preprocessing import MinMaxScaler

dummy = train[['fine_amount', 'admin_fee', 'state_fee', 'late_fee','discount_amount',
               'clean_up_cost', 'judgment_amount', 'payment_amount', 
               'balance_due','len_violator','len_description']]
dummy = MinMaxScaler(feature_range=(0,5)).fit_transform(dummy)
dummy = pd.DataFrame(data=dummy)
dummy.columns = ['fine_amount_N', 'admin_fee_N', 'state_fee_N', 'late_fee_N','discount_amount_N',
               'clean_up_cost_N', 'judgment_amount_N', 'payment_amount_N', 
               'balance_due_N','len_violator_N','len_description_N']
dummy
```

```python
train = pd.concat([train,dummy],axis=1)
train.drop(['fine_amount', 'admin_fee', 'state_fee', 'late_fee','discount_amount',
               'clean_up_cost', 'judgment_amount', 'payment_amount', 
               'balance_due','len_violator','len_description'],axis=1,inplace=True)
train.columns
```
